```python
import nltk
import random
from nltk.tokenize import word_tokenize

def calculate(operation, num1, num2):
    if operation == 'add' or operation == 'sum' :
        return num1 + num2
    elif operation == 'subtract' or operation == 'difference':
        return num1 - num2
    elif operation == 'multiply' or operation == 'product':
        return num1 * num2
    elif operation == 'divide' or operation == 'division':
        if num2 == 0:
            return None
        else:
            return num1 / num2
    else:
        return None
greetings = ['Hi, I am a calculator chatbot. How can I help you today?', 
             'Hello, I am a chatbot designed to perform calculations. What can I do for you?', 
             'Greetings! I am a calculator chatbot. How can I assist you?', 
             'Hey there! I am a chatbot designed to help you with your calculations. What do you need?', 
             'Salutations! I am a calculator chatbot. What calculation would you like me to perform?']
# Farewells
farewells = ['Goodbye! Have a nice day!', 
             'It was great chatting with you. Bye!', 
             'Until next time! Goodbye.', 
             'See you later!']
# Conversational prompts
prompts = ['How are you?', 
           'What else can you do?', 
           'Do you like math?', 
           'What is your favorite color?', 
           'What is your favorite movie?', 
           'What is your favorite food?']
print('Bot: ' + random.choice(greetings))

while True:
    user_input = input('User: ')
    tokens = word_tokenize(user_input)
    operation = None
    num1 = None
    num2 = None
    for token in tokens:
        if token.isdigit():
            if num1 is None:
                num1 = int(token)
            elif num2 is None:
                num2 = int(token)
                break
        elif token.lower() in ['add', 'subtract', 'multiply','product','sum','difference','division','divide']:
            operation = token.lower()
    if operation is not None and num1 is not None and num2 is not None:
        result = calculate(operation, num1, num2)
        if result is not None:
            print('Bot: ' + str(result))
        else:
            print('Bot: Sorry, I could not understand your input.')
    else:
        prompt = random.choice(prompts)
        if 'how are you' in user_input.lower():
            print('Bot: I\'m just a chatbot, so I don\'t have feelings, but thanks for asking! How are you?')
        elif 'tell me a joke' in user_input.lower():
            print('Bot: Why was the math book sad? Because it had too many problems!')
        elif 'fine' in user_input.lower():
            print('Bot: Oh hope you take care!')
        elif 'great' in user_input.lower():
            print('Bot: Happy to hear!')
        elif 'goodbye' in user_input.lower() or 'bye' in user_input.lower():
            print('Bot: ' + random.choice(farewells))
            break
        else:
            print('Bot: ' + prompt)
```
